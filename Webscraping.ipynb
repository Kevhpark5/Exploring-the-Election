```python
url = 'http://transcripts.cnn.com/TRANSCRIPTS/1911/11/se.01.html'
r = requests.get(url)
db = client.thlo1
pages = db.pages
pages.insert_one({'html': r.content})
soup1 = BeautifulSoup(r.content, 'html')
```

```python
div = soup1.find("div", {'style':'padding-left:10px;'})
p = div.findAll("p")[5]
```

```python
all_rows = []
empty_row = {
    'Speaker':None, 'Line':None
}

for i, row in enumerate(p):
    string = str(row)
    new_row = copy.copy(empty_row)
    a = ''
    if string[:7] == 'BURNETT':
        a = 'burnett'
        b = string[10:].lower()
    elif string[:5] == 'BIDEN':
        a = string[:5].lower()
        b = string[8:].lower()
    elif string[:8] == 'QUESTION':
        a = string[:8].lower()
        b = string[11:].lower()
    else:
        b = string.lower()
        
    new_row['Speaker'] = a
    new_row['Line'] = b
    all_rows.append(new_row)
    
for i in all_rows:
    if i['Speaker'] == '':
        if i['Line'] == '<br/>':
            all_rows.remove(i)
        elif i['Line'] == ' ':
            all_rows.remove(i)

for row in all_rows[1:-1]:
    pages.insert_one(row)
    
rows = pages.find()
th1 = pd.DataFrame(list(rows))

th1 = th1[th1['Line'] != ' '][1:-1]
lst = []
for idx, i in enumerate(th1['Speaker']):
    if i != '':
        a = i
        lst.append(a)
    elif i == '':
        lst.append(a)
        
th1['Speaker'] = lst
th1 = th1.drop(['_id', 'html'], axis = 1)
th1
```

```python
div = soup.find("div", {'style':'padding-left:10px;'})
p = div.findAll("p")[5]

all_rows = []
empty_row = {
    'Speaker':None, 'Line':None
}

for i, row in enumerate(p):
    string = str(row)
    new_row = copy.copy(empty_row)
    a = ''
    if string[:5] == 'LEMON':
        a = 'speaker'
        b = string[7:].lower()
    elif string[:9] == 'KLOBUCHAR':
        a = 'klobuchar'
        b = string[11:].lower()
    elif string[:8] == 'QUESTION':
        a = string[:8].lower()
        b = string[10:].lower()
    else:
        b = string.lower()
        
    new_row['Speaker'] = a
    new_row['Line'] = b
    all_rows.append(new_row)
    
for i in all_rows:
    if i['Speaker'] == '':
        if i['Line'] == '<br/>':
            all_rows.remove(i)
        elif i['Line'] == ' ':
            all_rows.remove(i)

for row in all_rows[:-1]:
    pages.insert_one(row)
    
rows = pages.find()
klobu5 = pd.DataFrame(list(rows))

klobu5 = klobu5[klobu5['Line'] != ' '][1:-1]
lst = []
for idx, i in enumerate(klobu5['Speaker']):
    if i != '':
        a = i
        lst.append(a)
    elif i == '':
        lst.append(a)
        
klobu5['Speaker'] = lst
klobu5 = klobu5.drop(['_id', 'html'], axis = 1)
klobu5
```
